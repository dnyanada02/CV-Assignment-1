### Importing Libraries

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from skimage.util import random_noise
from skimage import util
import matplotlib.pyplot as plt
from PIL import Image
```

### Uploading image

```python
from google.colab import files
uploaded = files.upload()

image = cv2.imread("flower.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
cv2_imshow(image)
```

### Adding salt and pepper noise using random noise function.

```python
img = np.array(image)
noisy1 = util.random_noise(img, mode='s&p', amount=0.05)
noisy2 = util.random_noise(img, mode='s&p', amount=0.1)
noisy3 = util.random_noise(img, mode='s&p', amount=0.2)
plt.subplot(221)
plt.title('original')
plt.xticks([]) # remove ticks
plt.yticks([])
plt.imshow(image)
plt.subplot(222)
plt.title('amount=0.05')
plt.xticks([]) # remove ticks
plt.yticks([])
plt.imshow(noisy1)
plt.subplot(223)
plt.title('amount=0.1')
plt.xticks([]) # remove ticks
plt.yticks([])
plt.imshow(noisy2)
plt.subplot(224)
plt.title('amount=0.2')
plt.xticks([]) # remove ticks
plt.yticks([])
plt.imshow(noisy3)
plt.show()
```

```python
noise = np.array(255*noisy3, dtype = 'uint8')
```

## Removing noise using Median Filter

For Kernel size = 3

```python
# Apply median filter
median = cv2.medianBlur(noise,3)
 
# Display the image
cv2_imshow(noise)
cv2_imshow(median)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

For Kernel size = 5

```python
# Apply median filter
median1 = cv2.medianBlur(noise,5)
 
# Display the image
cv2_imshow(noise)
cv2_imshow(median1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

 For Kernel size = 7

```python
# Apply median filter
median2 = cv2.medianBlur(noise,7)
 
# Display the image
cv2_imshow(noise)
cv2_imshow(median2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Applying Gaussian Filter 





For Kernel size = 3

```python
gausBlur = cv2.GaussianBlur(noise, (3,3),0) 
cv2_imshow(noise)
cv2_imshow(gausBlur)
```

For Kernel size = 5

```python
gausBlur1 = cv2.GaussianBlur(noise, (5,5),0) 
cv2_imshow(noise)
cv2_imshow(gausBlur)
```

For Kernel size = 7

```python
gausBlur2 = cv2.GaussianBlur(noise, (7,7),0) 
cv2_imshow(noise)
cv2_imshow(gausBlur2)
```
